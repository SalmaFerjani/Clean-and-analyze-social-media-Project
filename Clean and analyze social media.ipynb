### Import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

### Load the data into a Pandas DataFrame and Explore the data

```python
#Create a list containing some categories of socail media posts we can see in real life
list_category = ['Sport','Politcs','Fashion']

data_diction = {"Date":[],"Category":[],"Likes":[]}

for i in pd.date_range('2019-01-01',periods=20):
    for j in list_category:
        data_diction['Date'].append(i)
        data_diction['Category'].append(j)
        data_diction['Likes'].append(np.random.randint(0, 10000))
df = pd.DataFrame(data_diction)
display(df)
```

```python
likes = df.groupby(by = ['Date','Category']).sum().values
print(likes)
```

```python
df_likes = pd.DataFrame(likes.reshape(20, 3), index=range(1, 21), columns=['Fashion', 'Politics', 'Sport'])
print(df_likes)
```

### Histogramme des Likes par catégorie

```python
# Configuration des graphiques
plt.figure(figsize=(15, 10))

# Histogramme des Likes par catégorie
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='Likes', hue='Category', multiple='stack', palette='Set2')
plt.title('Histogramme des Likes par catégorie')
```

### Graphique en barres des Likes par date

```python
# Graphique en barres des Likes par date
plt.figure(figsize=(60, 25))
plt.subplot(2, 2, 2)
df_grouped = df.groupby('Date').sum().reset_index()
sns.barplot(data=df_grouped, x='Date', y='Likes', palette='viridis')
plt.xticks(rotation=90)
plt.title('Graphique en barres des Likes par date')
```

### Graphique de séries temporelles des Likes au fil du temps

```python
# Graphique de séries temporelles des Likes au fil du temps
plt.figure(figsize=(60, 25))
plt.subplot(2, 2, 3)
sns.lineplot(data=df, x='Date', y='Likes', hue='Category', palette='Set1')
plt.title('Graphique de séries temporelles des Likes')
```

```python
# Boxplot des Likes par catégorie
plt.figure(figsize=(40, 25))
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='Category', y='Likes', palette='Pastel1')
plt.title('Boxplot des Likes par catégorie')

plt.tight_layout()
plt.show()
```
